## **Grafica de barras**

```python
mun_nl = pd.DataFrame({'municipio': ['Monterrey', 'Guadalupe', 'Apodaca', 'San Nicol√°s'], 
                       'nacimientos':[500,630,768,400]})
mun_nl
```

```python
plt.bar(mun_nl['municipio'],
       mun_nl['nacimientos'])
plt.show()
```

```python
#Para cambiar el color de esta gr√°fica:
plt.bar(mun_nl['municipio'],
       mun_nl['nacimientos'],
       color = ['blue', 'green', 'red', 'yellow'])
plt.title('Nacimientos por municipio') #agregar titulo
plt.xlabel('Municipios de NL') #nombre al eje x
plt.ylabel('Cantidad de nacimientos') #nombre al eje y
plt. yticks([0,200,400,600,800]) #cambiar los saltos a escala y
plt.show()
```

```python
#Para cambiar el orden de las barras (mayor a menor):
mun_nl = mun_nl.sort_values('nacimientos', ascending= False)
plt.bar(mun_nl['municipio'],
       mun_nl['nacimientos'],
       color = ['blue', 'green', 'red', 'yellow'])
plt.title('Nacimientos por municipio') #agregar titulo
plt.xlabel('Municipios de NL') #nombre al eje x
plt.ylabel('Cantidad de nacimientos') #nombre al eje y
plt. yticks([0,200,400,600,800]) #cambiar los saltos a escala y
plt.show()
```

```python
#para poner varias graficas y ordenar; fig es nombre del lienzo
fig, (panel1, panel2)=plt.subplots(1,2) #subgraficas
fig.tight_layout # para que ambas graficas tengan el mismo tama√±o y diferente escala de ser necesario

# Panel1 (subgrafica 1)
x1 = x #eje x de grafico 1
y1 = y #eje y de grafico 1
panel1.plot(x,y,label = 'nombre de linea')
panel1.title.set_text('Grafica 1')

# Panel2 (subgrafica 1)
x2 = mun_nl['municipio'] #eje x de grafico 1
y2 = mun_nl['nacimientos'] #eje y de grafico 1
panel2.bar(x2,y2)
panel2.title.set_text('Grafica 2')

plt.show
```
