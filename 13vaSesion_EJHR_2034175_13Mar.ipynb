```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from statsmodels.tsa.api import seasonal_decompose
```

```python
pasajeros = pd.read_csv('https://raw.githubusercontent.com/jimmyzac/Estadistica-Aplicada-FCFM-UANL/main/bases_datos/airline_passengers.csv', index_col ='Month', parse_dates= True)
```

```python
pasajeros.rename(columns = {'Thousands of Passangers': 'Miles de Pasajeros'})
```

```python
pasajeros.plot(figsize=(12,6))
```

Serie estacional,va creciente donde el pico ocurre en vacaciones de verano por logica

```python
serie1 = seasonal_decompose(pasajeros['Thousands of Passengers'], model= 'mul')
serie1.plot()
```

```python
# modelo tren tenemos tendencia y ciclo VAMOS a quitar componente estacional
pasajeros['Series desestacionalizada'] = pasajeros['Thousands of Passengers'] / serie1.seasonal
```

```python
pasajeros['Series desestacionalizada'].plot();
```
